```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Angenommen, df ist Ihr DataFrame
df = pd.read_csv("./data_set/smartphones.csv") 

# Auswahl der quantitativen Merkmale
features = ['Battery', 'Memory', 'Primary_Storage', 'External_Storage', 'Display_Size', 'Display_Resolution', 'Display_Refresh_Rate', 'Primary_Camera', 'Front_Camera']

# Konvertiere die Daten zu numerischen Werten und behandle Fehler
for feature in features:
    df[feature] = pd.to_numeric(df[feature], errors='coerce')

# Berechnung der Pearson-Korrelation
correlation_matrix = df[features].corr()

# Erstellung einer Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Korrelationsmatrix der Merkmale')
plt.show()
```
