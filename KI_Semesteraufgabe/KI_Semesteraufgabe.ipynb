```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Funktion zur Umwandlung der Aufl철sung in Pixel
def resolution_to_pixels(resolution):
    if isinstance(resolution, str):
        width, height = resolution.split(' x ')
        return int(width) * int(height)
    else:
        return None

# Angenommen, df ist Ihr DataFrame
df = pd.read_csv("./data_set/smartphones.csv") 

# Anwenden der Funktion auf die 'Display_Resolution' Spalte und Speichern als neue Spalte
df['Display_Resolution_Pixels'] = df['Display_Resolution'].apply(resolution_to_pixels)

# Aktualisiere 'features' um die neue Spalte mit einzubeziehen und entferne die alte 'Display_Resolution' Spalte
features = ['Battery', 'Memory', 'Primary_Storage', 'Display_Resolution_Pixels', 'Display_Refresh_Rate', 'Primary_Camera', 'Front_Camera']

# Konvertiere die Daten zu numerischen Werten und behandle Fehler
for feature in features:
    df[feature] = pd.to_numeric(df[feature], errors='coerce')

# Berechnung der Pearson-Korrelation
correlation_matrix = df[features].corr()

# Erstellung einer Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Korrelationsmatrix der Merkmale')
plt.show()
```
